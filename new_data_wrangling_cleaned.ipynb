```python
import pandas as pd
import numpy as np
```

```python
df_dataset = pd.read_csv('movies_data.csv')
df_dataset.head()
```

```python
df_dataset.info()
```

```python
df_dataset['Year'] = pd.to_numeric(df_dataset['Year'], errors='coerce')
df_dataset['Box Office'] = df_dataset['Box Office'].replace('[\$,]', '', regex=True).astype(float)
df_dataset.head()
```

```python
missing = df_dataset.isna().sum()

missing_pct = (
    df_dataset.isna().mean() * 100
).round(2)

pd.DataFrame({
    "Missing Count": missing,
    "Missing %": missing_pct
}).sort_values("Missing %", ascending=False)
```

```python
df_dataset = df_dataset.dropna(
    subset=[
        "IMDb_ID",
        "Primary Genre",
        "Genre",
        "Year"
    ]
)
```

```python
df_dataset["IMDb Votes"] = (
    df_dataset["IMDb Votes"]
      .astype(str)
      .str.replace(",", "", regex=False)
)

df_dataset["IMDb Votes"] = pd.to_numeric(
    df_dataset["IMDb Votes"],
    errors="coerce"
)
```

```python
print("Revenue = 0:", (df_dataset["Revenue"] == 0).sum())
print("Budget = 0:", (df_dataset["Budget"] == 0).sum())
```

```python
df_dataset["Revenue"] = df_dataset["Revenue"].replace(0, np.nan)
df_dataset["Budget"] = df_dataset["Budget"].replace(0, np.nan)
```

```python
df_dataset[["Revenue", "Budget"]].isna().sum()
```

```python
current_year = 2026
start_year = current_year - 26

df_dataset_last_26_years_clean = df_dataset.loc[(df_dataset["Year"] >= start_year) & (df_dataset["Year"] <= current_year)].copy()

df_dataset_last_26_years_clean["Year"] = df_dataset_last_26_years_clean["Year"].astype(int)

print(f"Year Range       : {start_year} to {current_year}")
print(f"Total Records     : {len(df_dataset)}")
print(f"Filtered Records  : {len(df_dataset_last_26_years_clean)}")
df_dataset_last_26_years_clean.head()
```

```python
df_dataset_last_26_years_clean.to_csv("data_movies_cleaned.csv", index=False)
```
